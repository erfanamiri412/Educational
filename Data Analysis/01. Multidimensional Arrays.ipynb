```python
import numpy as np
```

```python
# 1D array
arr = np.array([10, 20, 30])
arr, arr.ndim, arr.shape, arr.size, arr.itemsize, arr.dtype
```

```python
# 2D array
arr2 = np.array([
    [10, 20, 30, 40],
    [50, 60, 70, 80]
])
arr2, arr2.ndim, arr2.shape, arr2.size, arr2.itemsize, arr2.dtype
```

```python
# 3D array
arr3 = np.array([
    [[10, 20], [30, 40]],
    [[50, 60], [70, 80]]
])
arr3 , arr3.ndim, arr3.shape, arr3.size, arr3.itemsize, arr3.dtype
```
